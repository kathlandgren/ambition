## Generate optimal threshold data

```python
## Define parameters
num_pts=1000
tmax_vec=np.linspace(1.01,1000,num_pts)
xvec=np.linspace(-3,6,3*num_pts)
reward_vec=np.zeros((len(xvec),len(tmax_vec)))
optimum_vec=np.zeros(len(tmax_vec))
```

```python
## iterate over tmax
for t, tmax in enumerate(tmax_vec):
    for i, thresh in enumerate(xvec):
        reward_vec[i,t]=a.estimated_cumulative_reward_sigma(thresh,tmax,1)

    
for t in range(len(tmax_vec)):
    # print(max(reward_vec[:,t]))
    optimum_vec[t]=xvec[np.argmax(reward_vec[:,t])] #optimal threshold
    # print(np.argmax(reward_vec[:,t]))
  
```

```python
# Save tmax_vec to a pickle file
with open('data/tmax_vec.pkl', 'wb') as f:
    pickle.dump(tmax_vec, f)

# Save optimum_vec to a pickle file
with open('data/optimum_vec.pkl', 'wb') as f:
    pickle.dump(optimum_vec, f)
```
